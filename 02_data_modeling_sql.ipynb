```python
# нормалізація назв колонок
df.columns = (
    df.columns
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df = df.rename(columns={
    'humidity(%)': 'humidity_pct',
    'temperature(c)': 'temperature_c',
    'wind_chill(c)': 'wind_chill_c',
    'distance(km)': 'distance_km',
    'visibility(km)': 'visibility_km',
    'wind_speed(kmh)': 'wind_speed_kmh',
    'precipitation(mm)': 'precipitation_mm',
    'pressure(hpa)': 'pressure_hpa'
})
```

```python
df.columns
```

```python
# cтворення основної таблиці accidents_main
accidents_main = df[[
    'id',
    'severity',
    'start_time',
    'end_time',
    'distance_km'
]]
```

```python
# cтворення таблиці location_data
location_data = df[[
    'id',
    'start_lat',
    'start_lng',
    'street',
    'city',
    'county',
    'state',
    'timezone'
]]
```

```python
# cтворення таблиці weather_data
weather_data = df[[
    'id',
    'weather_condition',
    'temperature_c',
    'wind_chill_c',
    'wind_direction',
    'wind_speed_kmh',
    'visibility_km',
    'precipitation_mm',
    'pressure_hpa',
    'humidity_pct'
]]
```

```python
# cтворення таблиці road_conditions
road_conditions = df[[
    'id',
    'amenity',
    'bump',
    'crossing',
    'give_way',
    'junction',
    'no_exit',
    'railway',
    'roundabout',
    'station',
    'stop',
    'traffic_calming',
    'traffic_signal'
]]
```

```python
# cтворення таблиці time_features
time_features = df[[
    'id',
    'sunrise_sunset',
    'hour',
    'day',
    'month',
    'year',
    'weekday',
    'is_weekend'
]]
```

```python
# експорт таблиць у CSV для подальшої роботи

accidents_main.to_csv(
    '/Users/veronika/Documents/US_Accidents_Project/data/sql_tables/accidents_main.csv',
    index=False
)

location_data.to_csv(
    '/Users/veronika/Documents/US_Accidents_Project/data/sql_tables/location_data.csv',
    index=False
)

weather_data.to_csv(
    '/Users/veronika/Documents/US_Accidents_Project/data/sql_tables/weather_data.csv',
    index=False
)

road_conditions.to_csv(
    '/Users/veronika/Documents/US_Accidents_Project/data/sql_tables/road_conditions.csv',
    index=False
)

time_features.to_csv(
    '/Users/veronika/Documents/US_Accidents_Project/data/sql_tables/time_features.csv',
    index=False
)
```
